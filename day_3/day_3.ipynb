```python
data = []
for idx, line in enumerate(input_.split("\n")):
    matches = list(re.finditer(r"[*@#\- =/\+%$&]", line))
    matches = [[idx,] + list(m.span()) + [m.group(),] for m in matches]
    data += matches
points_df = pd.DataFrame(data=data, columns=["y", "x_start", "x_end", "value"])
```

```python
points_df["geometry"] = points_df.apply(lambda row: Point(row["x_start"], row["y"]), axis=1)
```

```python
points_df = gpd.GeoDataFrame(points_df)
```

```python
points_df["geometry"] = points_df["geometry"].buffer(1.5)
```

```python
ax = numbers_df.plot()
points_df.plot(ax=ax, color="red", alpha=0.3);
```

```python
numbers_df.overlay(points_df, keep_geom_type=False)\
    .drop_duplicates(subset=['y_1', 'x_start_1', 'x_end_1', 'value_1'])\
    ["value_1"].astype(int).sum()
```

# Part 2

```python
overlay = numbers_df.overlay(points_df, keep_geom_type=False)
gears = overlay[overlay["value_2"] == "*"]\
    .sort_values(["y_2", "x_start_2", "y_1", "x_start_1"])\
    .groupby(["y_2", "x_start_2", "x_end_2", "value_2"], as_index=False)\
    .agg({"y_1": "size", "value_1": lambda s: s.astype(int).product()})\
    .rename(columns={"y_1": "size"})
gears = gears[gears["size"] == 2]
gears
```
